### Questão 1 — Criando e Explorando uma Series

Uma loja de jogos registrou as vendas (em unidades) de 5 consoles no último mês:

| Console | Vendas |
|---------|--------|
| PS5 | 120 |
| Xbox | 95 |
| Switch | 150 |
| Steam Deck | 60 |
| ROG Ally | 35 |

**a)** Crie uma Series chamada `vendas` com esses dados (use os nomes dos consoles como labels).

**b)** Exiba o valor de vendas do **Switch**.

**c)** Exiba a **soma total** de vendas e a **média** de vendas.

```python
# Seu código aqui
labels = ['PS5', 'Xbox', 'Switch', 'Steam Deck', 'ROG Ally']
dados = [120, 95, 150, 60, 35]

vendas = pd.Series(index=labels, data=dados)
print(vendas)
print(f'Soma : {vendas.sum()}')
print(f'Media : {vendas.mean()}')


```

### Questão 2 — Filtragem com Condicionais

Usando a Series `vendas` criada na Questão 1:

**a)** Filtre e exiba apenas os consoles que venderam **mais de 90 unidades**.

**b)** Filtre e exiba os consoles que venderam **acima da média**.

```python
# Seu código aqui
print(vendas[vendas > 90])
print("-"*70)
print(vendas[vendas > vendas.mean()])

```

### Questão 3 — Operações entre Series

No mês seguinte, as vendas foram:

| Console | Vendas |
|---------|--------|
| PS5 | 130 |
| Xbox | 88 |
| Switch | 145 |
| Steam Deck | 75 |
| ROG Ally | 50 |

**a)** Crie uma nova Series `vendas_mes2` com esses valores.

**b)** Calcule a **variação** de vendas entre o mês 2 e o mês 1 (mês 2 − mês 1).

**c)** Exiba apenas os consoles que **cresceram** em vendas.

```python
# Seu código aqui
labels_m2 = ['PS5', 'Xbox', 'Switch', 'Steam Deck', 'ROG Ally']
dados_m2 = [130, 88, 145, 75, 50]
vendas_mes2 = pd.Series(index=labels_m2, data=dados_m2)
print(vendas_mes2)
print("-"*70)
varia_vendas = vendas_mes2 - vendas
print(f'os produtos que aumentaram em vendas:\n{varia_vendas[varia_vendas > 0]} ')  
```

### Questão 4 — Criando e Acessando um DataFrame

Execute a célula abaixo para criar o DataFrame de referência:

```python
# Criando o DataFrame de referência
np.random.seed(10)
df = pd.DataFrame(
    index=['A', 'B', 'C', 'D', 'E'],
    columns=['W', 'X', 'Y', 'Z'],
    data=np.random.randint(1, 50, [5, 4])
)
print(df)
```

Usando o DataFrame `df` acima:

**a)** Selecione e exiba apenas as colunas **W** e **Z**.

**b)** Usando `loc()`, selecione as linhas **B** e **C** com as colunas **X** e **Y**.

**c)** Usando `iloc()`, selecione as **3 primeiras linhas** e as **2 últimas colunas**.

```python
# Seu código aqui
print(df[['W', 'Z']])
print("-"*70)
print(df.loc[['B', 'C'], ['X', 'Y']])
print("-"*70)
print(df.iloc[0:3, -2:])
```

### Questão 5 — Filtragem no DataFrame

Ainda usando o DataFrame `df`:

**a)** Exiba todas as linhas onde o valor da coluna **W** é **maior que 20**.

**b)** Qual é a **média da coluna Y** considerando apenas as linhas onde **Z > 10**?

```python
# Seu código aqui
maiores = df['W']
print(maiores[maiores > 20])
media_y = df[df['Z'] > 10]['Y'].mean()

print(f"A media da coluna Y para Z > 10 eh: {media_y:.2f}")
```
