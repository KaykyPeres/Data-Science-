## Series

### Questão 1
Crie duas Series com os seguintes valores:
- `seriesAno1`: `{'Java': 16.25, 'C': 16.04, 'Python': 9.85}`
- `seriesAno2`: `{'C': 16.21, 'Python': 12.12, 'Java': 11.68}`

```python
# Seu código aqui
dados1 = {'Java': 16.25, 'C': 16.04, 'Python': 9.85}
dados2 = {'C': 16.21, 'Python': 12.12, 'Java': 11.68}

seriesAno1 = pd.Series(dados1)
seriesAno2 = pd.Series(dados2)

print("Serie Ano 1:\n", seriesAno1)
print("\nSérie Ano 2:\n", seriesAno2)
```

### Questão 2
Os valores das Series criadas na Questão 1 representam as fatias de mercado (porcentagem) de 3 linguagens de programação populares em dois anos consecutivos. Para cada ano, apresente a **porcentagem total** que elas juntas representam no mercado.

```python
# Seu código aqui
total1 = seriesAno1.sum()
total2 = seriesAno2.sum()

print(f"Porcentagem total mercado - Ano 1: {total1}%")
print(f"Porcentagem total mercado - Ano 2: {total2}%")
```

### Questão 3
Apresente o **crescimento/declínio** no mercado de cada linguagem do primeiro ano para o segundo ano.

```python
# Seu código aqui
variacao = seriesAno2 - seriesAno1

print("Variação de mercado por linguagem:")
print(variacao)
```

### Questão 4
Baseado nos resultados da Questão 3, mostre **apenas os dados das linguagens que tiveram crescimento**.

```python
# Seu código aqui
crescimento = variacao[variacao > 0]

print("Linguagens que tiveram crescimento:")
print(crescimento)
```

### Questão 5
Se estas porcentagens de crescimento/declínio se mantivessem iguais para os próximos **2 anos**, qual seria a linguagem mais popular?

> **Dica:** use o método `nlargest(1)` no final para retornar rapidamente a label e o maior valor de uma Series.

```python
# Seu código aqui
projecao = seriesAno2 + (variacao * 2)
mais_popular = projecao.nlargest(1)

print("Projeção para daqui a 2 anos:")
print(projecao)
print("\nA linguagem que seria a mais popular e:")
print(mais_popular)
```
