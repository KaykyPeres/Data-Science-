### Questão 5
Faça uma função que receba a **taxa de mortalidade** de cada país (`Deathrate`) e retorne:
- O texto `'Balanced'` caso o valor seja **< 9**;
- O texto `'Urgent'` caso contrário.

Em seguida, crie um campo no Dataset chamado `'Humanitarian Help'` que receba estes valores para cada país. No final, mostre o Dataset para verificar se a inserção da nova coluna foi feita com sucesso.

```python
# Seu código aqui
def categorizar_ajuda(valor):
    if valor < 9:
        return 'Balanced'
    else:
        return 'Urgent'
    
df['Humanitarian Help'] = df['Deathrate'].apply(categorizar_ajuda)
print(df.head())

df.to_csv('paises_V2.csv', sep=';')
```
