Todos os exercícios utilizam o dataset **`cidades_brasil.csv`**, que contém informações sobre 25 cidades brasileiras.

```
Índice | Coluna      | Descrição
-------|-------------|----------------------------------
  0    | cidade      | Nome da cidade
  1    | estado      | Sigla do estado
  2    | populacao   | População estimada
  3    | area_km2    | Área em km²
  4    | idh         | Índice de Desenvolvimento Humano
  5    | regiao      | Região do Brasil
  6    | fundacao    | Ano de fundação
```

````python
# Importação e carregamento do dataset
import numpy as np

dados = np.loadtxt('cidades_brasil.csv', delimiter=',', dtype=str, skiprows=1)

# Colunas textuais
cidades    = dados[:, 0]
estados    = dados[:, 1]
regioes    = dados[:, 5]

# Colunas numéricas
populacao  = dados[:, 2].astype(int)
area       = dados[:, 3].astype(int)
idh        = dados[:, 4].astype(float)
fundacao   = dados[:, 6].astype(int)

print(f"Dataset carregado: {len(cidades)} cidades")
print(dados)
````

### Exercício 2 – Operações Estatísticas

**a)** Calcule e exiba a **média**, o **mínimo** e o **máximo** da população. Mostre também o **nome** da cidade mais populosa e da menos populosa (use `argmax` e `argmin`).

Dica: Caso necessário, para converter dados numéricos que estão como texto no array, você pode fazer a chamada da função `populacao  = dados[:, 2].astype(int)` colocando a tipagem no parêntezes (Ex: `int`, `float`)

**b)** Calcule e exiba a **soma total** das áreas de todas as cidades e a **média** do IDH do dataset.

**c)** Encontre o **índice** da cidade mais antiga (menor ano de fundação) e o **índice** da mais recente. Exiba os nomes e anos correspondentes.

````python
# Resolução do Exercício 2

print(f"Media: {populacao.mean()} ")
print(f"Minimo: {populacao.min()}")
print(f"Maximo: {populacao.max()}")
idx_max = populacao.argmax()
idx_min = populacao.argmin()
print(f"Cidade mais populosa: {cidades[idx_max]}")
print(f"Cidade menos populosa: {cidades[idx_min]}")
print("_"*70)

soma_total = np.sum(area)
print(f"a soma total da area : {soma_total}\n")
media_idh = np.sum(idh)/idh.size
print(f"A media do idg das cidades : {media_idh}")
print("_"*70)

mais_antiga_idx = fundacao.argmax()
menos_antiga_idx = fundacao.argmin()

print(f"a cidade mais antiga : {cidades[mais_antiga_idx]} fundada no ano {fundacao.min()}")
print(f"a cidade mais nova : {cidades[menos_antiga_idx]} fundada no ano {fundacao.max()}")

````

### Exercício 3 – Operações entre Arrays e Broadcasting

**a)** Calcule a **densidade demográfica** (população / área) de cada cidade e armazene em um novo array. Exiba as 5 cidades com **maior** densidade (exiba nome e valor).

**b)** Calcule a **idade** de cada cidade (2025 - ano de fundação) usando broadcasting. Exiba o array resultante.

**c)** Crie um array de **1's** com o mesmo tamanho que `idh`. Multiplique-o por **1000** e depois **some** com o array `idh * 1000` (IDH em escala de milésimos). Exiba o resultado.

````python
# Resolução do Exercício 3
densidade_demografica = populacao / area
arr = np.array(np.argsort(densidade_demografica)[-5:])
print(f"cidades com maior densidade demografica {cidades[arr]}")
print("_"*70)

idade_das_cidades = 2025 - fundacao
print(f"Idade das cidades : {idade_das_cidades}")
print("_"*70)

array_de_uns = np.ones(idh.size*1000)
idh_milesimos = array_de_uns  + (idh.size * 1000) 
print(f"IDH em escala de milesimos: {idh_milesimos}")
````

### Exercício 4 – Concatenação e Ordenação

**a)** Separe o array `populacao` em dois: um com as **10 primeiras** cidades e outro com as **10 últimas**. Concatene-os em um único array e exiba.

**b)** Ordene o array `idh` em ordem crescente usando `np.sort` e exiba. Qual é o menor e o maior IDH após a ordenação?

**c)** Use `np.argsort` no array `populacao` para obter os **índices** que ordenariam as cidades por população. Exiba os nomes das **5 cidades menos populosas** usando esses índices.

````python
# Resolução do Exercício 4
populacao1 = populacao[:10]
populacao2 = populacao[-10:]
print(populacao1)
print(populacao2)
concatenados = np.concatenate((populacao1 , populacao2))
print(f"O novo array com 20 cidades: {concatenados}")
print("_"*70)

idh_ordenado = np.sort(idh)
maior_idh = idh_ordenado.argmax()
menor_idh = idh_ordenado.argmin()
print(f"A cidade com maior IDH e : {cidades[maior_idh]} com {idh_ordenado.max()}")
print(f"A cidade com menor IDH e : {cidades[menor_idh]} com {idh_ordenado.min()}")
print("_"*70)

populacao_idx = np.argsort(populacao)[:5]
print(f"as 5 cidades menos populosas são :{cidades[populacao_idx]}")
````

### Exercício 5 – Reshape e Matrizes

**a)** Crie uma **matriz numérica** empilhando `populacao`, `area` e `fundacao` como colunas (use `np.column_stack`). Exiba as propriedades `shape`, `ndim` e `size`.

**c)** Selecione as **5 primeiras linhas** e as **2 primeiras colunas** da matriz (fatiamento). Exiba o resultado e os nomes das cidades correspondentes.
