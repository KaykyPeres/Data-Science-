## Exercício 1 - Manipulação de Strings

Crie um programa que leia seu nome completo e mostre:
* Seu nome com todas as letras maiúsculas
* Seu nome com todas as letras minúsculas
* Quantas letras ao todo tem seu nome
* E como seria se trocássemos seu último nome para "do Inatel"

```python
nomeCompleto = input("Nome completo: ").strip()

print(f"Maiusculas: {nomeCompleto.upper()}")
print(f"Minusculas: {nomeCompleto.lower()}")

tamanhoDoNome = len(nomeCompleto) - nomeCompleto.count(' ')
print(f"Seu nome tem {tamanhoDoNome} letras!")

lista = nomeCompleto.split()
lista[-1] = "do Inatel"
nomeInatel = " ".join(lista)
print(f"Nome modificado: {nomeInatel}")
```

## Exercício 2 - Tabuada com Intervalo

Mostre a tabuada de um número que o usuário escolher dentro de um intervalo específico também escolhido por ele.

```python
class Tabuada:
    def __init__(self, numero):
        self.numero = numero

    def mostrarTabuada(self, inicio, fim):
        print(f"\nTabuada do {self.numero} no intervalo de {inicio} a {fim}:")
        for i in range(inicio, fim + 1):
            resultado = self.numero * i
            print(f"{self.numero} x {i} = {resultado}")

num_escolhido = int(input("Qual numero você quer a tabuada? "))
comeco = int(input("Inicio do intervalo: "))
parada = int(input("Fim do intervalo: "))
tabuadaPersonalizada = Tabuada(num_escolhido)

tabuadaPersonalizada.mostrarTabuada(comeco, parada)
```

## Exercício 3 - Validação de Entrada

Faça um programa que leia o sexo de uma pessoa e diga se ela é homem (caso seja digitado M) ou mulher (caso seja digitado F). Caso seja digitado algo inválido, continue perguntando até que o usuário entre com um sexo válido.

```python
sexo = input("Digite o sexo M/F: ").strip().upper()

while sexo != "M" and sexo != "F":
    print("Desculpe! genero nao encontrado! digite novamente!")  
    sexo = input("Digite o sexo M/F: ").strip().upper()

if sexo == "M":
    print("Você é Homem!")
else:
    print("Você é Mulher! ")
```

## Exercício 4 - Cálculo de Passagem

Desenvolva um script que pergunte a distância de uma viagem em Km. Calcule o preço da passagem, cobrando R$0.50 por Km para viagens até 200Km e R$0.45 para viagens mais longas.

```python
distancia = float(input("Digite a distancia da sua viagem:"))

if distancia <= 200:
    preco = distancia * 0.50
elif distancia > 200:
    preco = distancia * 0.45

print(f"O preco da passagem ficou em R${preco:.2f}")
```

## Exercício 5 - Decomposição de Número

Faça um programa que leia um número entre 1000 e 9999 e mostre na tela:
* Qual o número da unidade
* Número da dezena
* Número da centena
* E número do milhar

```python
num = int(input("Entre com um numero entre 1000 a 9999 de preferencia"))

unidade = num // 1 % 10
dezena = num // 10 % 10
centena = num // 100 % 10
milhar = num // 1000 % 10

print(f"Milhar: {milhar}")
print(f"Centena: {centena}")
print(f"Dezena: {dezena}")
print(f"Unidade: {unidade}")
```

## Exercício 6 - Funções Matemáticas

Peça ao usuário para entrar com um número decimal. Em seguida, aplique e mostre o resultado:
* Da raiz quadrada deste número
* Função teto
* Função chão
* Sua parte inteira

```python
import math

num = float(input("entre com um valor decimal: "))

print(f"A raiz quadrada de {num} e {math.sqrt(num):.3f} ")
print(f"A funcao teto de {num} e {math.ceil(num)} ")
print(f"A funcao chao de {num} e {math.floor(num)} ")
print(f"A parte inteira de {num} e {math.trunc(num)}")

```
