## Exercício 3

Faça um programa que leia o nome e a média de um aluno e guarde-os em um dicionário. Em seguida, a partir da média (para ser aprovado deve ter média >=50), gere a situação final do aluno ('AP' ou 'RP'), que também deve ser guardada neste dicionário. No final, mostre todo o conteúdo deste dicionário;

```python
# Exercício 3

# Dicionário do aluno
alunos = [
    {
        "nome": "Kayky",
        "marticula": "177",
        "media": 100
    },
    {
        "nome": "joao cleber",
        "marticula": "256",
        "media": 47
    },
    {
        "nome": "Roberta",
        "marticula": "1322",
        "media": 72
    },
    {
        "nome": "Cleiton",
        "marticula": "122",
        "media": 59
    }
]

for aluno in alunos:
    if aluno["media"] >= 50:
        aluno["situacao"] = "AP"
    else:
        aluno["situacao"] = "RP"


# Leitura do nome
for aluno in alunos:
    print(f"Nome: {aluno["nome"]}")
print("-----------------------------------")

# Leitura da média
for aluno in alunos:
    print(f"Nome: {aluno["nome"]} - Media do aluno: {aluno["media"]}")
print("-----------------------------------")


# Determinar a situação final
for aluno in alunos:
    print(f"Nome: {aluno["nome"]} - Situação: {aluno["situacao"]}")
print("-----------------------------------")


# Mostrar todo o conteúdo do dicionário
print(alunos)
```

## Exercício 5

Desenvolva um programa que leia o nome, idade e sexo de n pessoas. No final, mostre:

a. A média de idade do grupo;

b. Quantas mulheres têm menos de 20 anos.

**Dica:** em Python, os operadores booleanos básicos são `and`, `or` e `not`.

```python
# Exercício 5

# Lista para armazenar as pessoas
pessoas = []

n = int(input("Digite o numero de pesoas a armazenar: "))
for i in range(n):
    nome = input("Digite o nome: ")
    idade = int(input("Digite a idade: "))
    sexo = input("Digite o genero com F ou M: ").upper()
    while sexo != "M" and sexo != "F":
        print("Desculpe! genero nao encontrado! digite novamente!")  
        sexo = input("Digite o sexo M/F: ").strip().upper()
    
    pessoas.append({"nome": nome, "idade": idade,"genero":sexo})


# a. Calcular a média de idade do grupo
soma_idades = 0
for pessoa in pessoas:
    soma_idades += pessoa["idade"]

media = soma_idades / n


# b. Contar quantas mulheres têm menos de 20 anos
contador = 0
for pessoa in pessoas:
    if pessoa["genero"] == "F" and pessoa["idade"] < 20:
        contador += 1


# Mostrar os resultados
print("Média de idade:", media)
print("Mulheres com menos de 20 anos:", contador)
```

## Exercício 6: Análise de Vendas - Projeto de Ciência de Dados

Você recebeu um dataset de vendas de uma loja online e precisa realizar uma análise completa dos dados. O dataset contém problemas de qualidade que precisam ser tratados.
**OBJETIVO:** Limpar o dataset e extrair insights valiosos sobre vendas, clientes, produtos e regiões usando apenas coleções Python!

Seja criativo e capture informaçÕes que você achar relevante!

### DICAS PARA O EXERCÍCIO:

**Limpeza de Dados:**
- Você pode usar `.strip()` para remover espaços extras e `.title()` para padronizar capitalização
- Você pode usar `abs()` para converter valores negativos ou simplesmente remover registros inválidos (valor/quantidade ≤ 0)
- Você pode criar um dicionário de mapeamento para unificar variações de categorias (ex: `'eletronicos' → 'Eletrônicos'`)
- Você pode trabalhar com cópias (`.copy()`) para preservar o dataset original

**Análises Possíveis:**
- Você pode usar **sets** para identificar valores únicos (clientes, produtos, regiões distintas)
- Você pode usar **dicionários** para agrupar e agregar dados (vendas por categoria, por região, por cliente)
- Você pode usar `sum()` e `len()` para calcular totais e contar elementos
- Você pode usar `max()` e `min()` com `key=lambda` para encontrar extremos (maior venda, melhor cliente)
- Você pode usar `sorted()` com `reverse=True` para criar rankings
- Você pode usar operações de conjuntos como `&` (interseção) e `|` (união) para análises comparativas


```python
# Dataset original
vendas = [
    {'id': 1, 'produto': 'Notebook', 'categoria': 'Eletrônicos', 'valor': 3500, 'quantidade': 2, 'regiao': 'Sudeste', 'cliente': 'João Silva'},
    {'id': 2, 'produto': 'Mouse', 'categoria': 'eletronicos', 'valor': 45, 'quantidade': 5, 'regiao': 'sul', 'cliente': 'Maria Santos'},
    {'id': 3, 'produto': 'Teclado', 'categoria': 'Eletrônicos', 'valor': 120, 'quantidade': 3, 'regiao': 'Sudeste', 'cliente': 'joão silva'},
    {'id': 4, 'produto': 'Cadeira Gamer', 'categoria': 'Móveis', 'valor': 890, 'quantidade': 1, 'regiao': 'Norte', 'cliente': 'Pedro Costa'},
    {'id': 5, 'produto': 'Mesa', 'categoria': 'moveis', 'valor': 450, 'quantidade': 2, 'regiao': 'Sul', 'cliente': 'Ana Lima'},
    {'id': 6, 'produto': 'Monitor', 'categoria': 'Eletrônicos', 'valor': 850, 'quantidade': 2, 'regiao': 'Sudeste', 'cliente': 'Carlos Mendes'},
    {'id': 7, 'produto': 'Webcam', 'categoria': 'Eletrônicos', 'valor': -250, 'quantidade': 4, 'regiao': 'Norte', 'cliente': 'Pedro Costa'},
    {'id': 8, 'produto': 'Cadeira Escritório', 'categoria': 'Móveis', 'valor': 350, 'quantidade': 3, 'regiao': 'Sudeste', 'cliente': 'JOAO SILVA'},
    {'id': 9, 'produto': 'Headset', 'categoria': 'eletronicos', 'valor': 180, 'quantidade': 0, 'regiao': 'Sul', 'cliente': 'Maria Santos'},
    {'id': 10, 'produto': 'SSD', 'categoria': 'Eletrônicos', 'valor': 380, 'quantidade': 3, 'regiao': 'norte', 'cliente': 'Fernanda Rocha'},
    {'id': 11, 'produto': '  Mouse Gamer  ', 'categoria': 'Eletrônicos', 'valor': 150, 'quantidade': 2, 'regiao': '  Sudeste  ', 'cliente': 'João Silva'},
    {'id': 12, 'produto': 'Teclado Mecânico', 'categoria': 'Eletronicos', 'valor': 0, 'quantidade': 1, 'regiao': 'Sul', 'cliente': 'maria santos'},
]

import unicodedata
import matplotlib.pyplot as plt

def limpar_texto(texto):
    texto = texto.strip()
    texto = texto.lower()
    texto = ''.join(
        c for c in unicodedata.normalize('NFD', texto)
        if unicodedata.category(c) != 'Mn'
    )
    texto = texto.title()
    return texto

vendas_tratadas = []
for venda in vendas:
    nova_venda = venda.copy()

    nova_venda['produto'] = limpar_texto(nova_venda['produto'])
    nova_venda['categoria'] = limpar_texto(nova_venda['categoria'])
    nova_venda['regiao'] = limpar_texto(nova_venda['regiao'])
    nova_venda['cliente'] = limpar_texto(nova_venda['cliente'])

    if nova_venda['valor'] < 0:
        nova_venda['valor'] = abs(nova_venda['valor'])
        
    nova_venda['valida'] = nova_venda['valor'] > 0 and nova_venda['quantidade'] > 0
    nova_venda['faturamento'] = nova_venda['valor'] * nova_venda['quantidade']

    vendas_tratadas.append(nova_venda)

# Calculos de faturamento(validos e nao validos) e percentual de impacto 
faturamento_total = sum(v['faturamento'] for v in vendas_tratadas)
faturamento_valido = sum(v['faturamento'] for v in vendas_tratadas if v['valida'])
faturamento_invalido = faturamento_total - faturamento_valido

if faturamento_total > 0:
    percentual_perda = (faturamento_invalido / faturamento_total) * 100
else:
    percentual_perda = 0

# Regiao com mais vendas invalidas 
invalido_por_regiao = {}

for v in vendas_tratadas:
    if not v['valida']:
        regiao = v['regiao']
        invalido_por_regiao[regiao] = invalido_por_regiao.get(regiao, 0) + 1

print('\nVendas invalidas por região:')
for regiao, qtd in invalido_por_regiao.items():
    print(f'{regiao}: {qtd}')

# Cliente que gerou mais faturamento 
faturamento_por_cliente = {}

for v in vendas_tratadas:
    if v['valida']:
        cliente = v['cliente']
        faturamento_por_cliente[cliente] = faturamento_por_cliente.get(cliente, 0) + v['faturamento']

cliente_top = max(faturamento_por_cliente, key=faturamento_por_cliente.get)

print('\nCliente com maior faturamento:')
print(cliente_top, '-> R$', f'{faturamento_por_cliente[cliente_top]:,.2f}')

# Ranking de clientes por faturamento
ranking_clientes = sorted(faturamento_por_cliente.items(),key=lambda p: p[1], reverse=True)

print('\nRanking de clientes:')
for cliente, valor in ranking_clientes:
    print(cliente, '-> R$', f'{valor:,.2f}')


print('='*70)
print('ANÁLISE DE VENDAS - PROJETO DE CIÊNCIA DE DADOS')
print('='*70)
print(f'\nDataset original: {len(vendas)} registros')
print(f'Dataset tratado: {len(vendas_tratadas)} registros')

print('\n' + '='*70)
print('ANÁLISES DETALHADAS')
print('='*70)
print('\n' + '='*70)
print('RESUMO FINANCEIRO')
print('='*70)
print(f'Faturamento Total: R$ {faturamento_total:,.2f}')
print(f'Faturamento Valido: R$ {faturamento_valido:,.2f}')
print(f'Faturamento Invalido: R$ {faturamento_invalido:,.2f}')
print(f'Impacto percentual das vendas invalidas: {percentual_perda:.2f}%')
print('='*70)

# graficozinho com matplot so pra ficar bala! 
nomes = [item[0] for item in ranking_clientes[:5]]
valores = [item[1] for item in ranking_clientes[:5]]

plt.figure(figsize=(10, 6))
bars = plt.bar(nomes, valores, color='skyblue', edgecolor='navy')

plt.title('Top 5 Clientes por Faturamento', fontsize=14, fontweight='bold')
plt.xlabel('Clientes', fontsize=12)
plt.ylabel('Faturamento (R$)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 100, f'R$ {yval:,.2f}', 
             ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()
```
