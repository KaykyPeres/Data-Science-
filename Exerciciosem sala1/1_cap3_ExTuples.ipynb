### ExTuples ‚Äì Q1: M√∫sicas favoritas

1. Crie uma tupla chamada `musicas_favoritas` com **5 m√∫sicas** que voc√™ mais gosta
2. Exiba a primeira e a √∫ltima m√∫sica usando √≠ndices
3. Exiba as 3 m√∫sicas do meio usando *slicing*
4. Mostre o tamanho da tupla com `len()`

````python
# ExTuples ‚Äì Q1
# Crie sua tupla com 5 m√∫sicas favoritas
musicas_favoritas = ('Bleed the freak','shine on you crazy diamond','Break Stuff','faint','snuff')

# Primeira m√∫sica
print(musicas_favoritas[0])

# √öltima m√∫sica
print(musicas_favoritas[4])

# As 3 m√∫sicas do meio (slicing)
print(musicas_favoritas[1:4])

# Tamanho da tupla
print(len(musicas_favoritas))
````

### ExTuples ‚Äì Q2: Filmes preferidos

1. Crie uma tupla chamada `filmes` com **4 filmes** que voc√™ recomendaria a um amigo
2. Exiba os filmes em **ordem alfab√©tica** com `sorted()`
3. Descubra em qual √≠ndice est√° o seu favorito com `.index()`
4. Itere com `for` imprimindo cada filme no formato: `"Recomendo: <filme>"`

````python
# ExTuples ‚Äì Q2
# Crie sua tupla com 4 filmes favoritos
filmes = ('Bastardos inglorios','Duna pt2','Senhor dos aneis o retorno do rei','O lobo de wall street')

# Filmes em ordem alfab√©tica
print(sorted(filmes))

# √çndice do seu filme favorito
print(filmes[3])

# Itera√ß√£o com mensagem
for i, filmes in enumerate(filmes):
    print(f'filme {i}: {filmes}')
````

### ExTuples ‚Äì Q3: Notas do semestre

1. Crie duas tuplas: `notas_p1` e `notas_p2`, cada uma com **3 notas** (reais ou fict√≠cias) de mat√©rias diferentes
2. Concatene as duas em `todas_notas` usando `+`
3. Exiba a nota m√°xima e m√≠nima com `max()` e `min()`
4. Escolha uma nota e conte quantas vezes ela aparece com `.count()`

````python
# ExTuples ‚Äì Q3
# Crie as tuplas com suas notas
notas_p1 =(60,75,68) 
notas_p2 =(90,85,60) 

# Concatena√ß√£o
todas_notas = notas_p1 + notas_p2
print(todas_notas)

# Nota m√°xima e m√≠nima
print(max(todas_notas), min(todas_notas))

# Contagem de uma nota espec√≠fica
print(todas_notas.count(60))
````
