### ExSets ‚Äì Q1: G√™neros musicais em comum

1. Crie um set `meus_generos` com os g√™neros musicais que **voc√™** gosta
2. Crie um set `generos_amigo` com os g√™neros que **um amigo** gosta (compartilhe alguns em comum)
3. Descubra todos os g√™neros que pelo menos um de voc√™s gosta (**Uni√£o `|`**)
4. Descubra os g√™neros que **s√≥ voc√™** gosta (**Diferen√ßa `-`**)
5. Descubra os g√™neros que **ambos** gostam (**Interse√ß√£o `&`**)

````python
# ExSets ‚Äì Q1
# Crie seus sets com g√™neros musicais
meus_generos = {'Rock','indie','NuMetal','Blues'}
generos_amigo = {'Rock','NuMetal','Trap','sertaNOJO'}

# Uni√£o (todos os g√™neros)
todosGeneros = meus_generos | generos_amigo
print(todosGeneros)

# Diferen√ßa (s√≥ voc√™ gosta)
meusApenas = meus_generos - generos_amigo
print(meusApenas)

# Interse√ß√£o (ambos gostam)
emComum = meus_generos & generos_amigo
print(emComum)
````

### ExSets ‚Äì Q2: Limpando duplicatas do hist√≥rico

1. Crie uma **lista** chamada `historico_filmes` com pelo menos **8 filmes**, repetindo pelo menos **3 t√≠tulos** de prop√≥sito
2. Converta a lista para um **set** chamado `filmes_unicos` para eliminar as repeti√ß√µes
3. Exiba quantos filmes √∫nicos voc√™ tem com `len()`
4. Verifique se um filme espec√≠fico est√° no set usando `in`

````python
# ExSets ‚Äì Q2
# Crie sua lista com filmes repetidos de prop√≥sito
historico_filmes ={'senhor dos aneis','lobo de wall street','pulp fiction','mickey 17','senhor dos aneis','de volta para o futuro','mickey 17','alien o oitavo passageiro','lobo de wall street'} 

# Converter para set
filmes_unicos = historico_filmes
print(filmes_unicos)

# Quantidade de filmes √∫nicos
print(len(filmes_unicos))

# Verificar se um filme est√° no set
filme = "pulp fiction"

if filme in filmes_unicos:
    print("O filme est√° no set!")
else:
    print("O filme n√£o est√° no set.")
````

### ExSets ‚Äì Q3: Habilidades da equipe

1. Crie um set `minhas_skills` com suas habilidades t√©cnicas (linguagens, ferramentas, etc.)
2. Crie um set `skills_colega` com habilidades de um colega (compartilhe algumas)
3. Exiba habilidades que **s√≥ voc√™ tem** (Diferen√ßa)
4. Exiba habilidades que **s√≥ o colega tem** (Diferen√ßa inversa)
5. Exiba habilidades que **ambos dominam** (Interse√ß√£o)
6. Exiba o conjunto completo de habilidades da equipe (Uni√£o)

````python
# ExSets ‚Äì Q3
# Crie os sets com suas habilidades e as do colega
minhas_skills ={'python','java','c++','c#','unity','js','docker','kanban'}
skills_colega ={'python','java','c++','js','rust','docker','golang','android studio'}

# S√≥ voc√™ tem
print(minhas_skills.difference(skills_colega))

# S√≥ o colega tem
print(skills_colega.difference(minhas_skills))

# Ambos dominam
print (minhas_skills.intersection(skills_colega))

# Equipe completa
print(minhas_skills.union(skills_colega))
````
