### ExAll â€“ Q1: Biblioteca pessoal

Monte um mini sistema para organizar seus livros favoritos usando **todas as coleÃ§Ãµes**:

1. **Tupla** `categorias`: gÃªneros literÃ¡rios que vocÃª aprecia (ex: ficÃ§Ã£o, tecnologia, HQ...)
2. **Lista de dicionÃ¡rios** `biblioteca`: pelo menos **4 livros**, cada um com `titulo`, `autor`, `categoria`, `ano` e `lido` (True/False)
3. **Set** `autores_unicos`: gerado a partir da `biblioteca`, com todos os autores sem repetiÃ§Ã£o *(dica: itere sobre a lista e use `.add()` no set)*
4. Filtre e exiba apenas os livros que vocÃª **jÃ¡ leu** (`lido == True`)
5. Exiba o total de autores Ãºnicos e as categorias disponÃ­veis

```python
# ExAll â€“ Q1

# 1. Tupla de categorias
categorias = ("ficcao", "tecnologia", "hq", "fantasia","Filosofia")

# 2. Lista de dicionÃ¡rios (biblioteca)
biblioteca = [
    {
        "titulo": "Duna",
        "autor": "Frank Herbert",
        "categoria": "ficcao",
        "ano": 1965,
        "lido": True
    },
    {
        "titulo": "Clean Code",
        "autor": "Robert C. Martin",
        "categoria": "tecnologia",
        "ano": 2008,
        "lido": True
    },
    {
        "titulo": "Watchmen",
        "autor": "Alan Moore",
        "categoria": "hq",
        "ano": 1987,
        "lido": False
    },
    {
        "titulo": "O Hobbit",
        "autor": "J.R.R. Tolkien",
        "categoria": "fantasia",
        "ano": 1937,
        "lido": True
    }
]

# 3. Set de autores Ãºnicos
autores_unicos = set()
for livro in biblioteca:
    autores_unicos.add(livro["autor"])

# 4. Livros jÃ¡ lidos
print("Livros lidos:")
for livro in biblioteca:
    if livro["lido"] == True:
        print(livro["titulo"])

# 5. Total de autores Ãºnicos e categorias disponÃ­veis
print("Total de autores Ãºnicos:", len(autores_unicos))
print("Categorias disponÃ­veis:", categorias)
```

### ExAll â€“ Q2: Ranking de jogos

VocÃª e 2 amigos jogaram os mesmos games e deram notas de 1 a 10. Use **todas as coleÃ§Ãµes**:

1. **Tupla** `jogos`: nomes de **4 games** que vocÃªs jogaram
2. **DicionÃ¡rio** `notas_jogos`: chaves sÃ£o os nomes dos jogos e valores sÃ£o **listas** com as 3 notas (uma por jogador)
3. Calcule e exiba a **mÃ©dia** de cada jogo *(dica: use `sum()` e `len()`)*
4. Identifique qual jogo teve a **maior mÃ©dia**
5. Crie um **set** `jogos_perfeitos` com os jogos que receberam pelo menos uma nota **10**

```python
jogos = ("Silent Hill f", "Expedition 33", "Hollow Knight Silksong", "Final Fantasy 7 Remake")

notas_jogos = {
    "Silent Hill f": [8, 9, 7],
    "Expedition 33": [10, 9, 10],
    "Hollow Knight Silksong": [10, 10, 9],
    "Final Fantasy 7 Remake": [7, 6, 6]
}

medias = {}
for jogo, notas in notas_jogos.items():
    media = sum(notas) / len(notas)
    medias[jogo] = media
    print(f"{jogo} - media: {media:.2f}")

maior_jogo = max(medias, key=medias.get)
print("Maior media:", maior_jogo)

jogos_perfeitos = set()

for jogo, notas in notas_jogos.items():
    if 10 in notas:
        jogos_perfeitos.add(jogo)

print("Jogos com nota 10:", jogos_perfeitos)

```
