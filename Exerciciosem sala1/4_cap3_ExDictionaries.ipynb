### ExDictionaries ‚Äì Q1: Perfil pessoal

1. Crie um dicion√°rio chamado `meu_perfil` com as chaves: `nome`, `idade`, `cidade`, `curso` e `hobbie_favorito` ‚Äî preenchidas com seus dados reais
2. Acesse e exiba seu `nome` e `hobbie_favorito`
3. Adicione a chave `linguagem_favorita` com o valor correspondente
4. Atualize sua `idade` para um ano a mais
5. Exiba todas as **chaves** com `.keys()` e todos os **valores** com `.values()`

````python
# ExDictionaries ‚Äì Q1
# Crie seu perfil com seus dados reais
meu_perfil = {
    'nome':'kayky' ,
    'idade': 22,
    'cidade':'Sao Lourenco' ,
    'curso': 'GES',
    'hobbie_favorito':'Jogar no PC' ,
}

# Acessar nome e hobbie_favorito
print(meu_perfil['nome'],' /// ',meu_perfil['hobbie_favorito'])        

# Adicionar linguagem_favorita
meu_perfil['linguagemFav'] = 'Java'

# Atualizar idade
meu_perfil['idade'] = 23

# Exibir todas as chaves
print(meu_perfil.keys())

# Exibir todos os valores
print(meu_perfil.values())
````

### ExDictionaries ‚Äì Q2: Card√°pio do restaurante favorito

1. Crie um dicion√°rio chamado `cardapio` onde as **chaves** s√£o pratos do seu restaurante favorito e os **valores** s√£o os pre√ßos ‚Äî inclua pelo menos **5 pratos**
2. Exiba o pre√ßo de um prato espec√≠fico
3. Adicione um novo prato ao card√°pio
4. Remova um prato com `del`
5. Itere com `.items()` exibindo: `"Prato: X ‚Äì R$ Y"`

````python
# ExDictionaries ‚Äì Q2
# Crie o card√°pio do seu restaurante favorito
cardapio = {
    "Hamburguer": 25.90,
    "Pizza": 39.90,
    "Lasanha": 32.50,
    "Batata frita": 15.00,
    "Refrigerante": 7.00
}

# Exibir pre√ßo de um prato espec√≠fico
print(cardapio["Pizza"])

# Adicionar novo prato
cardapio['Raviolli'] = 29.90

# Remover um prato
del cardapio["Batata frita"]

# Iterar com .items()
for prato, preco in cardapio.items():
    print(f"Prato: {prato} - R$ {preco}")
````

### ExDictionaries ‚Äì Q3: Lista de contatos

**Sintaxe de lista de dicion√°rios:**
```python
contatos = [
    {'nome': 'Ana', 'cidade': 'SP'},
    {'nome': 'Bruno', 'cidade': 'RJ'},
]
```

1. Crie uma lista de dicion√°rios chamada `contatos` com **3 amigos ou colegas**, cada um com as chaves: `nome`, `telefone`, `email` e `cidade` (pode usar dados fict√≠cios)
2. Exiba o **nome** do primeiro contato: `contatos[0]['nome']`
3. Exiba o **email** do segundo contato
4. Itere sobre a lista exibindo nome e cidade de cada contato
5. Adicione a chave `aniversario` ao terceiro contato

````python
# ExDictionaries ‚Äì Q3
# Crie sua lista de contatos
contatos = [
    {
        "nome": "Ana",
        "telefone": "1199489-1597",
        "email": "ana@gmail.com",
        "cidade": "SP"
    },
    {
        "nome": "Bruno",
        "telefone": "2198888-2222",
        "email": "bruno@gmail.com",
        "cidade": "RJ"
    },
    {
        "nome": "jonatas",
        "telefone": "3597777-3333",
        "email": "jonatas@gmail.com",
        "cidade": "SRS"
    }
]

# Nome do primeiro contato
print(contatos[0]["nome"])

# Email do segundo contato
print(contatos[1]["email"])

# Iterar exibindo nome e cidade
for contato in contatos:
    print(f"Nome: {contato['nome']} - Cidade: {contato['cidade']}")

# Adicionar aniversario ao terceiro contato
contatos[2]["aniversario"] = "11/09"
print(contatos)
````
