### ExLists ‚Äì Q1: Lista de desejos

1. Crie uma lista chamada `lista_desejos` com **4 itens** que voc√™ quer comprar ou conquistar (game, viagem, gadget, curso...)
2. Adicione mais **2 itens** ao final com `append()`
3. Insira um item **na posi√ß√£o 2** com `insert()`
4. Remova o item que voc√™ menos quer com `remove()`
5. Exiba a lista final

````python
# ExLists ‚Äì Q1
# Crie sua lista de desejos com 4 itens
lista_desejos =['Body pillow','faqueira magnetica','sapateira shopee','capinha p/ analogico do controle'] 

# Adicionar 2 itens ao final
lista_desejos.append('travesseiro emma')
print(lista_desejos)

# Inserir item na posi√ß√£o 2
lista_desejos.insert(2,'capinha de celular verde')
print(lista_desejos)

# Remover o item que voc√™ menos quer
lista_desejos.remove('faqueira magnetica')
# del lista_desejos[1] tambem funcionaria 

# Exibir lista final
print(lista_desejos)
````

### ExLists ‚Äì Q2: Placar de um jogo

1. Crie uma lista chamada `placar` com **5 entradas** no formato `'Nome: Pontua√ß√£o'` (use nomes de amigos, personagens ou jogadores reais)
2. Ordene o placar em **ordem alfab√©tica** com `sort()`
3. Reverta a ordem com `sort(reverse=True)`
4. Substitua o jogador da posi√ß√£o `0` por um novo usando atribui√ß√£o direta
5. Exiba o placar final

````python
# ExLists ‚Äì Q2
# Crie seu placar com 5 jogadores
placar = ['jonas: 2','chris: 4','renzo: 1','rhuan: 5', 'mosca: 0']

# Ordenar alfabeticamente
placar.sort()
print(placar)

# Ordenar em ordem inversa
placar.sort(reverse=True)
print(placar)

# Substituir jogador da posi√ß√£o 0
placar[0] = "Marcelo: 4"

# Exibir placar final
print(placar)
````

### ExLists ‚Äì Q3: S√©ries para assistir

1. Crie uma lista chamada `series` com **6 s√©ries** que voc√™ quer ou j√° assistiu
2. Exiba as **3 primeiras** usando *slicing*
3. Remova a s√©rie da posi√ß√£o `1` com `pop()` e guarde o valor em uma vari√°vel
4. Delete a √∫ltima s√©rie com `del`
5. Exiba a s√©rie removida pelo `pop()` e a lista resultante

````python
# ExLists ‚Äì Q3
# Crie sua lista com 6 s√©ries
series = ['Breaking bad','GOT','One Piece','Black Mirror','Love death & Robots','Supernatural']

# As 3 primeiras s√©ries (slicing)
print(series[0:3])

# Remover s√©rie na posi√ß√£o 1 e guardar o valor
removida = series.pop(1)

# Deletar a √∫ltima s√©rie
del series[4]

# Exibir s√©rie removida e lista resultante
print('S√©rie removida:', removida)
print('Lista restante:', series)
````
