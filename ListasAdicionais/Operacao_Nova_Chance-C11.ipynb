### Questão 1 – Controle de Doações em Array

O albergue recebeu doações de alimentos (em kg) durante **10 dias consecutivos**. Use `np.random.seed(42)` e crie um array de **10 números aleatórios inteiros entre 5 e 50** representando os kg doados por dia.

**a)** Exiba as propriedades `size`, `ndim` e `shape` do array;  
**b)** Calcule e exiba a **média**, o **mínimo** e o **máximo** de doações diárias;  
**c)** Mostre o **índice do dia** com maior doação e o **índice do dia** com menor doação;  
**d)** Mostre quantos dias tiveram doação **acima da média**.

````python
# Questão 1
np.random.seed(42)
doacoes = np.random.randint(5, 51, size=10)

print("Array de doações (kg):", doacoes)

# a) Propriedades do array
print(doacoes.size)
print(doacoes.ndim)
print(doacoes.shape)

# b) Média, mínimo e máximo
print("Média:", doacoes.mean())
print("Mínimo:", doacoes.min())
print("Máximo:", doacoes.max())


# c) Índice do maior e menor dia de doação
print(f"Índice do maior dia de doação:{doacoes.argmax()}")
print(f"Índice do menor dia de doação: {doacoes.argmin()}")


# d) Quantidade de dias com doação acima da média
media_doacoes = doacoes.mean()
dias_acima_media = (doacoes > media_doacoes).sum()
print(f"Quantidade de dias com doação acima da média: {dias_acima_media}")
````

### Questão 2 – Mapa do Albergue e Seleção de Candidatos

**Parte A — Mapa do Albergue**

O gestor do albergue quer visualizar a ocupação dos **12 quartos** em formato de planta. Crie um array com os valores de **1 a 12** e:

**a)** Remodelar (`reshape`) para uma **matriz 3x4** (3 andares, 4 quartos por andar);  
**b)** Exibir a **soma de cada andar** (linha) e a **soma de cada coluna de quartos**.

---

**Parte B — Seleção para Capacitação**

O CRAS possui um programa de capacitação profissional. Para participar, o candidato precisa ter **IDH de bairro >= 0.65** e **ter menos de 45 anos**. Use os arrays abaixo:

```python
nomes    = np.array(["Zé", "Maria", "Cláudio", "Lurdes", "Expedito", "Raimundo", "Fátima", "Tonho"])
idades   = np.array([38, 52, 41, 29, 47, 33, 61, 44])
idh      = np.array([0.71, 0.58, 0.80, 0.63, 0.70, 0.75, 0.55, 0.68])
```

**c)** Filtre e exiba os **nomes dos candidatos aprovados** (IDH >= 0.65 **E** idade < 45);  
**d)** Exiba a **média de IDH** dos candidatos aprovados.

````python
# Questão 2 — Parte A
quartos = np.arange(1, 13)

# a) Reshape para matriz 3x4
quartos_mtz = quartos.reshape(3, 4)
print(quartos_mtz)


# b) Soma de cada andar (linhas) e cada coluna
soma_andares = quartos_mtz.sum(axis=1)
soma_colunas = quartos_mtz.sum(axis=0)

print("Soma de cada andar (linhas):", soma_andares)
print("Soma de cada coluna:", soma_colunas) 
````

````python
# Questão 2 — Parte B
nomes    = np.array(["Zé", "Maria", "Cláudio", "Lurdes", "Expedito", "Raimundo", "Fátima", "Tonho"])
idades   = np.array([38, 52, 41, 29, 47, 33, 61, 44])
idh      = np.array([0.71, 0.58, 0.80, 0.63, 0.70, 0.75, 0.55, 0.68])

# c) Nomes dos aprovados (IDH >= 0.65 E idade < 45)
aprovados = nomes[(idh >= 0.65) & (idades < 45)]
print("Nomes dos aprovados:", aprovados)


# d) Média de IDH dos aprovados
media_idh_aprovados = idh[ (idh >= 0.65) & (idades < 45) ].mean()
print(f"Média de IDH dos aprovados: {media_idh_aprovados:.2f}")
````

### Questão 4 – O Grande Censo dos Assistidos

O CRAS fez um censo completo e salvou os dados no arquivo `assistidos.csv`. Carregue o arquivo e responda:

**a)** Exiba **todos os dados** e as **informações gerais** do DataFrame (`.info()`);  
**b)** Calcule a **média de renda** dos assistidos da região `"Norte"`;  
**c)** Mostre o **nome e a renda** do assistido com a **maior renda**;  

````python
# Questão 4 — Carregamento do CSV
df = pd.read_csv("assistidos.csv")

# a) Imprima os dados e info
print(df)
print(df.info())
````

````python
# b) Média de renda da região Norte
media_norte = df[df['regiao'] == 'Norte']['renda'].mean()
print(f"Média de renda da região Norte: {media_norte:.2f}")
````

````python
# c) Nome e renda do assistido com maior renda
idx_maior = df['renda'].idxmax()
nome_top = df.loc[idx_maior, 'nome']
renda_top = df.loc[idx_maior, 'renda']

print(f"Maior renda: {nome_top} com R$ {renda_top:.2f}")
````
